### Q1: Code cell

Use the code cell below to produce your dataframe.

```python
import pandas as pd
import requests

URL = "https://api.nal.usda.gov/fdc/v1/foods/search"
API_KEY = "cH66HghcdeLPueZHNsGm9Uz8sYYMJysBbVr93kT2" #Insert your API key here

TO_REPLACE = {
    'Energy':	 	 	 	 	 				'Energy',
    'Carbohydrate, by difference':				'Carbs',
    'Protein':									'Protein',
    'Fiber, total dietary':						'Fiber',
    'Total lipid (fat)':						'Fat'
}

def food_nutrients(food):
  params = {
      "api_key": API_KEY,
      "query": food,
      "requireAllWords": True
  }

  res = requests.get(URL, params=params)
  food_data = res.json()
  nutrients_list = []

  for foods in food_data.get("foods"):
    food_nutrients = {"Food": foods}
    for nutrients in foods.get("foodNutrients"):
      name = nutrients.get("nutrientName")
      if name in TO_REPLACE:
        food_nutrients[TO_REPLACE[name]] = nutrients.get("value", 0)
    nutrients_list.append(food_nutrients)

  if nutrients_list:
      df = pd.DataFrame(nutrients_list).drop(columns=["Food"]).mean().to_frame().T
      df["Food"] = food
      return df

  return pd.DataFrame()


def menu_nutrients(foods):
    data = []
    for f in foods:
      df = food_nutrients(f)
      if not df.empty:
        data.append(df)


    big_df = pd.concat(data, ignore_index=True)
    big_df = big_df[["Food"] + list(TO_REPLACE.values())]
    big_df = big_df.sort_values(by="Food", ascending=True)
    big_df = big_df.reset_index(drop=True)
    big_df = big_df.set_index("Food")
    return big_df
```

Use the cell below to run your code with the required inputs.

```python
## This cell must show the output
required_foods = [
    "turkey",
    "macaroni and cheese",
    "mashed potatoes",
    "bread stuffing",
    "ham",
    "sweet potato souffle",
    "cranberry sauce",
    "mixed vegetables",
    "apple pie",
    "pecan pie"
]

menu_nutrients(required_foods)
```
